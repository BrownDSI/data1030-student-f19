Problem 1) One ML algorithm we didn't cover during class is the nearest neighbor algorithm. The principle behind nearest neighbor methods is to find a predefined number of training samples closest in distance to the new point, and predict the label from these nearest neighbors. The number of samples can be a user-defined constant (k-nearest neighbor learning), or vary based on the local density of points (distance-based neighbor learning). The distance can, in general, be any metric measure: standard Euclidean distance is the most common choice.

Read more about this method [here](https://scikit-learn.org/stable/modules/neighbors.html#nearest-neighbors-classification) and [here](https://scikit-learn.org/stable/modules/neighbors.html#nearest-neighbors-regression).

```python
from sklearn.neighbors import KNeighborsRegressor

import numpy as np
from sklearn.svm import SVR
np.random.seed(10)
def true_fun(X):
    return np.cos(1.5 * np.pi * X)

n_samples = 30

X = np.random.rand(n_samples)
y = true_fun(X) + np.random.randn(n_samples) * 0.1

X_new = np.linspace(-0.5, 1.5, 2000)
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 16})


plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.scatter(X,y,label='training data')
plt.plot(np.linspace(0, 1, 100),true_fun(np.linspace(0, 1, 100)),label='true function')
reg = KNeighborsRegressor(n_neighbors=1,weights='uniform')
reg.fit(X[:, np.newaxis],y)
y_new = reg.predict(X_new[:, np.newaxis])
plt.plot(X_new,y_new,'r',label='prediction')
plt.xlabel('x')
plt.ylabel('y')
plt.title('n_neighbors = 1')
plt.legend()

plt.subplot(2,2,2)
plt.scatter(X,y,label='training data')
plt.plot(np.linspace(0, 1, 100),true_fun(np.linspace(0, 1, 100)),label='true function')
reg = KNeighborsRegressor(n_neighbors=3,weights='uniform')
reg.fit(X[:, np.newaxis],y)
y_new = reg.predict(X_new[:, np.newaxis])
plt.plot(X_new,y_new,'r',label='prediction')
plt.xlabel('x')
plt.ylabel('y')
plt.title('n_neighbors = 3')
plt.legend()

plt.subplot(2,2,3)
plt.scatter(X,y,label='training data')
plt.plot(np.linspace(0, 1, 100),true_fun(np.linspace(0, 1, 100)),label='true function')
reg = KNeighborsRegressor(n_neighbors=10,weights='uniform')
reg.fit(X[:, np.newaxis],y)
y_new = reg.predict(X_new[:, np.newaxis])
plt.plot(X_new,y_new,'r',label='prediction')
plt.xlabel('x')
plt.ylabel('y')
plt.title('n_neighbors = 10')
plt.legend()


plt.subplot(2,2,4)
plt.scatter(X,y,label='training data')
plt.plot(np.linspace(0, 1, 100),true_fun(np.linspace(0, 1, 100)),label='true function')
reg = KNeighborsRegressor(n_neighbors=30,weights='uniform')
reg.fit(X[:, np.newaxis],y)
y_new = reg.predict(X_new[:, np.newaxis])
plt.plot(X_new,y_new,'r',label='prediction')
plt.xlabel('x')
plt.ylabel('y')
plt.title('n_neighbors = 30')
plt.legend()

plt.tight_layout()
plt.savefig('figures/kneighbors_uni_reg.png',dpi=300)
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
X = data.data
y = data.target
print(len(y))
print(np.shape(X))
ftr_names = data.feature_names
target_names = data.target_names
print(target_names)
print(ftr_names)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

def best_clf(X_train,y_train,X_test,y_test):
    
    best_test_scores = []
    
    # Lasso
    alpha = np.logspace(-1,3,5)

    train_acc = np.zeros([len(alpha)]) 
    test_acc = np.zeros([len(alpha)])
    for i in range(len(alpha)):
        lr = LogisticRegression(penalty='l1',C = 1e0/alpha[i],solver='saga',max_iter=1e4)
        lr.fit(X_train,y_train)
        train_acc[i] = accuracy_score(y_train,lr.predict(X_train))
        test_acc[i] = accuracy_score(y_test,lr.predict(X_test))

    print('log reg:',np.max(test_acc))
    indcs = np.unravel_index(np.argmax(test_acc),np.shape(test_acc))
    print(alpha[indcs[0]]) 
    best_test_scores.append(np.max(test_acc))
    
    # random forest
    max_depth = [1,2,3,5,10,15,20]
    min_samples_split = [2,4,8,16,32]
    RF_accs = []
    train_acc = np.zeros([len(max_depth),len(min_samples_split)]) 
    test_acc = np.zeros([len(max_depth),len(min_samples_split)])
    for i in range(len(max_depth)):
        for j in range(len(min_samples_split)):
            lr = RandomForestClassifier(min_samples_split = min_samples_split[j],max_depth=max_depth[i],n_estimators=100,random_state=0)
            lr.fit(X_train,y_train)
            train_acc[i,j] = accuracy_score(y_train,lr.predict(X_train))
            test_acc[i,j] = accuracy_score(y_test,lr.predict(X_test))
    print('RF:',np.max(test_acc))
    indcs = np.unravel_index(np.argmax(test_acc),np.shape(test_acc))
    print(max_depth[indcs[0]],min_samples_split[indcs[1]]) 
    best_test_scores.append(np.max(test_acc))
    
    # SVM
    C = np.logspace(-4,4,9)
    gamma = np.logspace(-4,4,9)
    train_acc = np.zeros([len(C),len(gamma)]) 
    test_acc = np.zeros([len(C),len(gamma)])
    for i in range(len(C)):
        for j in range(len(gamma)):
            lr = SVC(C = C[i],gamma = gamma[j],kernel='rbf')
            lr.fit(X_train,y_train)
            train_acc[i,j] = accuracy_score(y_train,lr.predict(X_train))
            test_acc[i,j] = accuracy_score(y_test,lr.predict(X_test))
    print('SVC:',np.max(test_acc)) 
    indcs = np.unravel_index(np.argmax(test_acc),np.shape(test_acc))
    print(C[indcs[0]],gamma[indcs[1]]) 
    best_test_scores.append(np.max(test_acc))
    
    # nearest neighbors
    n_neighbors = [1,3,10,30,100,300]

    train_acc = np.zeros([len(n_neighbors)]) 
    test_acc = np.zeros([len(n_neighbors)])
    for i in range(len(n_neighbors)):
        lr = KNeighborsClassifier(n_neighbors = n_neighbors[i])
        lr.fit(X_train,y_train)
        train_acc[i] = accuracy_score(y_train,lr.predict(X_train))
        test_acc[i] = accuracy_score(y_test,lr.predict(X_test))

    print('nearest neighbor:',np.max(test_acc))
    indcs = np.unravel_index(np.argmax(test_acc),np.shape(test_acc))
    print(n_neighbors[indcs[0]]) 
    best_test_scores.append(np.max(test_acc))
    
    return best_test_scores
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

print(best_clf(X_train,y_train,X_test,y_test))
```
